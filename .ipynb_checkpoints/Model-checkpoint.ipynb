```python
import pandas as pd
```

```python
df = pd.read_csv('advertising.csv')
df.head()
```

```python
print(df.shape)
print(df.info())
print(df.describe())
```

### There are no NA values in the dataset

```python
#Removing the unwanted column Instant
df.drop("instant", axis=1, inplace=True)
df.head()
```

```python
df['dteday'] = pd.to_datetime(df['dteday'], format='%d-%m-%Y')
df['day'] = df['dteday'].dt.day
```

```python
df.head()
```

```python
df['season'] = df['season'].apply(lambda x: '100' if x==1 else x)
df['season'] = df['season'].apply(lambda x: '010' if x==2 else x)
df['season'] = df['season'].apply(lambda x: '001' if x==3 else x)
df['season'] = df['season'].apply(lambda x: '000' if x==4 else x)
df.head()
```
